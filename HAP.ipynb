```python
JD_SKILLS = {
    "python",
    "retrieval",
    "ranking",
    "embeddings",
    "vector database",
    "faiss",
    "pinecone",
    "weaviate",
    "qdrant",
    "machine learning",
    "llm",
    "recommendation",
    "aws",
    "nlp"
}
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('all-MiniLM-L6-v2')

jd_description = (
    "We are looking for a Machine Learning Engineer with strong skills in retrieval, ranking, "
    "embeddings, and vector databases (Faiss, Pinecone, Weaviate, Qdrant). "
    "Experience with LLMs, recommendation systems, AWS, and NLP is highly valued. "
    "Proficiency in Python is essential."
)

jd_embedding = model.encode(jd_description)
```

```python
def skill_score(candidate):
    candidate_skills = set(
        skill["name"].lower() for skill in candidate["skills"]
    )
    matches = len(candidate_skills.intersection(JD_SKILLS))
    return matches / len(JD_SKILLS)
```

```python
score = skill_score(data[0])
print(score)
```

```python
def experience_score(candidate):

    years = candidate["profile"].get(
        "years_of_experience", 0
    )

    if years >= 6:
        return 1.0
    elif years >= 4:
        return 0.8
    elif years >= 2:
        return 0.6
    else:
        return 0.3
```

```python
print(experience_score(data[0]))
```

```python
def behavioral_score(candidate):

    signals = candidate["redrob_signals"]

    score = 0


    if signals.get("open_to_work_flag", False):
        score += 20


    score += signals.get(
        "profile_completeness_score", 0
    ) / 5


    score += signals.get(
        "recruiter_response_rate", 0
    ) * 20


    score += signals.get(
        "offer_acceptance_rate", 0
    ) * 20


    score += signals.get(
        "github_activity_score", 0
    )


    if signals.get("verified_email", False):
        score += 5

    if signals.get("verified_phone", False):
        score += 5

    return score / 100
```

```python
def quality_penalty(candidate):

    signals = candidate["redrob_signals"]

    penalty = 0

    if signals.get(
        "profile_completeness_score",0
    ) < 50:
        penalty += 0.15

    if not signals.get(
        "verified_email",False
    ):
        penalty += 0.05

    if not signals.get(
        "verified_phone",False
    ):
        penalty += 0.05

    return penalty
```

```python
def final_score(candidate):

    return (
        0.40 * skill_score(candidate) +
        0.25 * experience_score(candidate) +
        0.35 * behavioral_score(candidate)
    )

print(f"Final Score for candidate 0: {final_score(data[0])}")
```

```python
def semantic_score(candidate):

    text = ""

    text += candidate["profile"].get("summary", "")

    for skill in candidate["skills"]:
        text += " " + skill["name"]

    embedding = model.encode(text)

    score = cosine_similarity(
        [jd_embedding],
        [embedding]
    )[0][0]

    return float(score)
```

```python
def final_score(candidate):

    return (
        0.35 * semantic_score(candidate) +
        0.25 * skill_score(candidate) +
        0.15 * experience_score(candidate) +
        0.25 * behavioral_score(candidate) -
        quality_penalty(candidate)
    )
```

```python
all_candidates = data
```

```python
ranked_candidates = sorted(
    all_candidates,
    key=final_score,
    reverse=True
)

print("Top Candidates by Final Score (with Quality Penalty):")
for i, candidate in enumerate(ranked_candidates[:100], start=1):
    print(
        f"{i}. Candidate ID: {candidate['candidate_id']}, Final Score: {round(final_score(candidate), 4)}"
    )
```

```python
def explanation(candidate):

    return {
        "candidate_id":
            candidate["candidate_id"],

        "skill_score":
            round(skill_score(candidate),3),

        "experience_score":
            round(experience_score(candidate),3),

        "behavior_score":
            round(behavioral_score(candidate),3),

        "semantic_score":
            round(semantic_score(candidate),3)
    }

print(explanation(ranked_candidates[0]))
```

```python
import pandas as pd

submission_rows = []

for rank, candidate in enumerate(ranked_candidates[:100], start=1):

    score = round(final_score(candidate), 4)

    reasoning = (
        f"{candidate['profile'].get('current_title','Professional')} "
        f"with {candidate['profile'].get('years_of_experience',0):.1f} yrs experience; "
        f"matched key skills; strong behavioral signals"
    )

    submission_rows.append({
        "candidate_id": candidate["candidate_id"],
        "rank": rank,
        "score": score,
        "reasoning": reasoning
    })

submission = pd.DataFrame(submission_rows)

submission.to_csv(
    "/content/drive/MyDrive/submission.csv",
    index=False
)

print("submission.csv created successfully")
submission.head()
```
